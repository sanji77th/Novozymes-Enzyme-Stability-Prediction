```python
train_df = train_df.dropna(subset=['pH','data_source'])
```

```python
train_df = train_df.reset_index()
```

```python
train_df.info(null_counts=True)
```

```python
print(type(train_df['protein_sequence'][0]))
num_of_seq = []
seq_values = []
id = 0
for sequence in train_df['protein_sequence']:
    id += 1
    seq_total = 0
    count = len(sequence) + 1
    num_of_seq.append(len(sequence))
    for unit in sequence:
        count -= 1
        b = 91 - ord(unit)
        mul = count*b
        seq_total += mul
        #print(unit,b,count,mul)
    #print(id,seq_total)
    seq_values.append(seq_total)
print(seq_values)
```

```python
np_seq_values = np.array(seq_values)
np_num_of_seq = np.array(num_of_seq)
train_df['sequence_value'] = np_seq_values.tolist()
train_df['num_of_sequences'] = np_num_of_seq.tolist()
```

```python
protien_Alph = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', 'X', 'B', 'Z', 'J']

protien_lis = []

for i in range(len(protien_Alph)):
    protien_lis.append([])

test_str = train_df['protein_sequence'][0]  

for sequence in train_df['protein_sequence']:
    res = {}
  
    for keys in sequence:
        res[keys]=res.get(keys, 0)+1
    #print(res)

    for k in range(len(protien_Alph)):
        if protien_Alph[k] in res:
            #print(res[protien_Alph[k]])
            protien_lis[k].append(res[protien_Alph[k]])
        else:
            #print(0)
            protien_lis[k].append(0)
        
print(protien_lis)
```

```python
for k in range(len(protien_Alph)):
    train_df[protien_Alph[k]] = np.array(protien_lis[k])
```

```python
y = train_df['tm']
X = new_train_data.drop(['tm'], axis=1)
```

```python
X
```

```python
from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler()
scaleX = scaling.fit_transform(new_train_data[['pH','sequence_value','num_of_sequences','A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', 'X', 'B', 'Z', 'J']])
scaleX
```

```python
X = pd.DataFrame(scaleX, columns = ['pH','sequence_value','num_of_sequences','A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', 'X', 'B', 'Z', 'J'])
X
```

```python
y.to_frame()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 5)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
X_train
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```

```python
from sklearn.metrics import mean_absolute_error
mn_ab_error = mean_absolute_error(y_test, y_pred)
mn_ab_error
```

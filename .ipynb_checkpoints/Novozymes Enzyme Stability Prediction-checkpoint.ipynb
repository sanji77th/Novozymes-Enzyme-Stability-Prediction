```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv('Data/train.csv')
```

```python
train_df
```

```python
train_df.info()
```

```python
print(type(train_df['protein_sequence'][0]))
num_of_seq = []
seq_values = []
id = 0
for sequence in train_df['protein_sequence']:
    id += 1
    seq_total = 0
    count = len(sequence) + 1
    num_of_seq.append(len(sequence))
    for unit in sequence:
        count -= 1
        b = 91 - ord(unit)
        mul = count*b
        seq_total += mul
        #print(unit,b,count,mul)
    #print(id,seq_total)
    seq_values.append(seq_total)
print(seq_values)  
```

```python
np_seq_values = np.array(seq_values)
np_num_of_seq = np.array(num_of_seq)
train_df['sequence_value'] = np_seq_values.tolist()
train_df['num_of_sequences'] = np_num_of_seq.tolist()
```
